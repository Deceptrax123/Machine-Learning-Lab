```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("train.csv")
df
```

```python
X=df.drop(['id','target'],axis=1)
```

```python
X
```

```python
from sklearn.cluster import KMeans

kmeans=KMeans(n_clusters=3,random_state=0,max_iter=300,algorithm='lloyd').fit(X)
t=kmeans.labels_
```

```python
import matplotlib.pyplot as plt

plt.hist(t)
plt.title("Number of points on cluster")
```

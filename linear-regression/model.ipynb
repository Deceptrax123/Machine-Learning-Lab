```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train=pd.read_csv("./train.csv")
```

```python
train
```

```python
columns_drop=['case','Unnamed: 0']
train=train.drop(columns_drop,axis=1)
```

```python
train
```

```python
Y=train['totlngth']
```

```python
X=train.drop(['totlngth'],axis=1)
```

```python
X
```

```python
Y
```

```python
X=np.array(X)
Y=np.array(Y)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.10,shuffle=True)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression().fit(x_train,y_train)
```

```python
score=reg.score(x_train,y_train)
```

```python
score
```

```python
predicted=reg.predict(x_train)
```

```python
predicted
```

```python
from sklearn.metrics import mean_squared_error
mse_train=mean_squared_error(y_train,predicted)
```

```python
mse_train
```

```python
#check for test set
test_predicted=reg.predict(x_test)
```

```python
mse_test=mean_squared_error(y_test,test_predicted)
```

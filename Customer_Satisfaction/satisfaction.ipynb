```python
import pandas as pd
import numpy as np
```

```python
df_amazon = pd.read_csv("amazon_uk_shoes_products_dataset_2021_12.csv")
df_amazon.head()
```

```python
columns = ["product_name","review_title","review_rating"]
df1 = pd.DataFrame(df_amazon,columns=columns)
df1
```

```python
print(len(df1.product_name.unique()))
```

```python
review_type = []
for ind in df1.index:
    if df1["review_rating"][ind] >= 3:
        review_type.append(1)

    else:
        review_type.append(-1)

df1["review_type"] = review_type
df1
```

```python
df3=df1.groupby("product_name").head(3).reset_index(drop=True)
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("./dataset/data.csv")
```

```python
df
```

```python
#pre-processing look for nans
df.isna().sum().sum()

#Shows that there are 569 nan values that need to be filled up or dropped
```

```python
#removing columns with only NaN values
df=df.dropna(axis=1)
```

```python
df.isna().sum().sum()
```

```python
#The Un-named column with all nans have been dropped. There are no more null values
df
```

```python
#mapping categorical variables to numeric features
mapping={'M':1,'B':0}

df=df.replace(mapping)
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
#plotting correlation matrix for dimensionality reduction
plt.matshow(df.corr())
```

```python
df.corr()
```

```python
#From the correlation matrix, features with less than 0.05 dependency can be dropped.
drop_columns=['texture_se','fractal_dimension_mean','smoothness_se','symmetry_se']

df=df.drop(drop_columns,axis=1)
```

```python
df
```

```python
#Statistical Dataanlysis
df.describe()
```

```python
#Data Visualization
plt.hist(df['diagnosis'])
plt.title("Distribution of Benign and Malignant Tumors")
```

```python
#Scatter plot to depict corrleation between concave points and diagnosis
x=df['concave points_worst']
y=df['diagnosis']

plt.scatter(x,y)
plt.title("Relation between concave points and diagnosis")
```

```python
#Normalize around expectation of 0 and distribution standard deviation of 1.
for i in list(df.columns):
    if (i!='id' and i!='diagnosis'):
        df[i]=(df[i]-df[i].mean())/df[i].std()
```
